```python
response = req.get("https://catalog.data.gov/dataset?q=" + domain + " &sort=views_recent+desc&ext_location=&ext_bbox=&ext_prev_extent=-164.53125%2C-80.17871349622823%2C164.53125%2C80.17871349622823")
                   
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content)

print(soup.prettify())
```

```python
div_t = soup.find("div", class_="module-content")

print(div_t.prettify())
```

```python
list_items =[]
```

```python
list_items = div_t.find_all("li", class_="dataset-item has-organization")
```

```python
for i in list_items:
    print(i.prettify())
```

```python
# to auto - download the data files 
```

```python
download_list=[]

for i in range(0,4):
    #print(list_items[i])
    download_list.append(list_items[i].find("li").find("a").get("href"))
```

```python
for i in download_list:
    webbrowser.open(i)
```

```python

h3_list = []

for i in list_items:
    h3_list.append(i.find("h3"))
```

```python
h3_list
```

```python
a_list =[]

for i in h3_list:
    if i is None:
        None
    else:
        a_list.append(i.find("a"))
```

```python
a_list
```

```python
file_list=[]

for i in a_list:
    file_list.append(i.text)
```
