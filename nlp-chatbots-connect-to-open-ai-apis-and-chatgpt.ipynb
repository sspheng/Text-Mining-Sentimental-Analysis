```python
model_engine = "text-davinci-002"

prompt = "write story on World War 2"
```

```python
#https://platform.openai.com/docs/models/gpt-3-5
```

```python


response = openai.Completion.create(
    engine=model_engine,
    prompt=prompt,
    max_tokens=50   #tokens means words -- max words to generate
)
```

 we’re using the “text-davinci-002” model engine, which is one of the most advanced models available. 
 
 We’re also providing a prompt for the text generation, which is “Once upon a time, there was a “.

we’re setting the max_tokens parameter to 50, which limits the length of the generated text to 50 tokens (roughly equivalent to words).

```python

print(response.choices[0].text)
```
