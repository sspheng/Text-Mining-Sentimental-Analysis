```python
# Making a GET request
r = req.get('https://www.geeksforgeeks.org/python-programming-language/')
 
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
 
s = soup.find('div', class_='entry-content')
print(s.prettify())

```

```python
content = s.find_all('p')
 
print(content)
```

```python
# Finding by id
s = soup.find('div', id= 'main')
```

```python
r = req.get('https://www.geeksforgeeks.org/python-programming-language/')
 
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
 
s = soup.find('div', class_='entry-content')
 
lines = s.find_all('p')
lines
```

```python

for line in lines:
    print("0000", line)
```

```python
images_list = []
 
images = btf.select('img')

images
```

```python
for image in images:
    src = image.get('src')
    alt = image.get('alt')
    images_list.append({"src": src, "alt": alt})
     
```

```python
for image in images_list:
    print(image)
```
