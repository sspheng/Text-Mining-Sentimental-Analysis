* OBJECTIVE -- GSG (Gap Sentence Generation) +  MLM (Masked Language Model)

The authors propose a new Transformer based encoder-decoder pre-training objective called GSG (Gap Sentence Generation) in which they mask out a sentence in the document and then try to reproduce that sentence using the remaining sentences. 

* ROGUE Metric

Selection techs

* Random Selection  Randomly sampling k sentences from a list of all sentences in the input document.
* Leading Selection  Selecting top-k sentences from the starting of the document.
* Principal Selection  Selecting top-k sentences based on their score in the document. Scoring is done by calculating ROUGE-1 for sentence and input documents. This would ensure selecting the sentence that is very much representative of the entire document. Also, since the final summarization evaluation would be ROUGE based, so selecting masked sentences and maximizing on that makes sense. They did not mention anything about comparing other scoring metrics, but graph-based techniques such as LexRank, TextRank could have been tried.

The **tokens syntax in model.generate(**tokens) is known as "keyword argument unpacking" in Python

In this example, tokens is a dictionary containing the keys "input_ids" and "attention_mask" along with their respective values. 
Essentially, 

model.generate(input_ids=[1, 2, 3], attention_mask=[1, 1, 1]) 



is equivalent to 


model.generate(**tokens).

<s>: Start-of-sequence token.

Examples of special tokens include:

* [CLS]: Start-of-sequence token.
* [SEP]: Separator token used to mark the end of a sequence or the boundary between two sequences.
* [PAD]: Padding token used to make input sequences of equal length.
* [UNK]: Token used to represent unknown or out-of-vocabulary (OOV) words.
